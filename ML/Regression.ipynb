```python
import pandas as pd
from sklearn import linear_model
import numpy as np
from sklearn.metrics import r2_score
from sklearn.preprocessing import PolynomialFeatures 
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import metrics
import seaborn as seabornInstance
```

```python
mergedData = pd.read_csv("data/ML_Reg.csv");

#Splitting dataset into 80% training and 20% testing data
X = mergedData[['SP', 'P', 'O', 'AC', 'MD', 'CW', 'N', 'FA']].values
y = mergedData[['Probability']].values

plt.figure(figsize=(15,10))
plt.tight_layout()
seabornInstance.distplot(mergedData['Probability'])
#print(mergedData.describe())

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)



#Creating a regression model
regr = linear_model.LinearRegression(fit_intercept = True, n_jobs = -1)
regr.fit(X_train, y_train) 
 

#print(regr.coef_)

#Predicting dependent variables in testing dataset
y_pred = regr.predict(X_test)


# Testing the accuracy of the linear model

print(r2_score(y_test,y_pred))




```
